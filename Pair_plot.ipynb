# Pair Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np

var = sns.load_dataset("tips")
var
```

```python
sns.pairplot(var,hue="sex",hue_order=["Female","Male"],x_vars=["total_bill","tip"],kind="kde",diag_kind="hist",markers=["*",">"])

# eta shudhu numerical data niye kaj kore
# hue er maddhome amra kono kichu alada kore highlight korte pari. jemon ekhane amra male female alada kore dekhiyesi.
# ami jodi chai sob numerical column use korbo na. specific kicu use korbo then ami setao korte pari. amake vars=["column_name","column_name"] use korte hobe.
# ekhane var ekta user defined variable. amra jekono name dite pari. 
# hue_order diye amra sequence customize korte pari. 
# kon axis e amra ki dibo setao amra decide korte pari. Ex: x_vars=["total_bill","tip"]
# pairplot-এ kind প্যারামিটার ব্যবহার করে তুমি যে ধরনের প্লট দেখাতে চাও তা ঠিক করতে পারো। scatter(default), kde, reg, hist
# diagonal plot e ekeksomoy amra ekek plot dekhte pai. kokhono kde, kokhono hist. amra etake fixed jekono ekta use korte pari. 
# diag_kind="hist/kde"
# 







plt.show()
```
