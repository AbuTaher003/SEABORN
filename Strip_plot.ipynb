# Strip Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

var = sns.load_dataset("tips")
var
```

```python
# eta kichu ta box plot er moto karon ekhane data gula alada kore show kore. 
# eta kichu ta abar scatter plot er moto karon ekhane data gula dot dot hoye choriye chitiye thake.


sns.stripplot(x="day",y="total_bill",data=var,hue="sex",palette="pastel",marker=">")


# ekhane amra linewidth, edgecolor, size, marker, alpha ei parameter gula use korte pari.
# strip plot e amra male female er jonno dictionary use kore alada alada kore marker use korte pari na. but eta manually kora jay.

#------------------------------------------------------------------------------------------------------------------------------------------------------------
# # Male এর জন্য প্লট
# sns.stripplot(x="day", y="total_bill", data=var[var["sex"] == "Male"], marker=m["Male"], color="blue", label="Male")

# Female এর জন্য প্লট
# sns.stripplot(x="day", y="total_bill", data=var[var["sex"] == "Female"], marker=m["Female"], color="pink", label="Female")
#------------------------------------------------------------------------------------------------------------------------------------------------------------

plt.legend(loc="upper left") # label er position change kore dilam.
plt.show()
```
