# Scatter Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

var = sns.load_dataset("penguins")
var
```

```python
sns.scatterplot(x="bill_length_mm",y="flipper_length_mm",data=var,hue="sex",style="sex",
                size="sex",sizes=(100,60),palette="Accent",alpha=1,markers={"Male":'o',"Female":'*'})

# hue="sex" use kore male female alada kore indicate kore dilam and ekta automatic label add hoyeche.
# style="sex" use kore amra style change korte pari.
# size="sex" use korle automatic size change hobe.
# nijer iccha moto size dite arekta parameter use korte hobe seta holo sizes=(), ekhane male female re size onujayi fix kora jay.
# palette="Accent" use kore color change kora jay. google e onek colour pawa jabe.
# alpha use kore jhapsa kora jay. er range 0 to 1.
# markers={"Male":'o',"Female":'*'} use kore iccha moto mark kora jay. eta dictionary te korte hoy.




plt.show()
```
