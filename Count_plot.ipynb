# Count Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

var = sns.load_dataset("tips")
var
```

```python
# amra barplot , histogram plot diye ja korte pari sob e count plot use kore korte pari. but count plot er subidha holo ekhane sob kichu ekdom hisab kore dey graph e automatic. So this is better than bar and histogram.



sns.countplot(x="sex",data=var,hue="smoker",palette="bwr")

# ekhane x axis e sex niyechi. and y axis e count automatic count hoyeche. 
# amra ekhane graph customize korte iccha moto saturation, color, alpha parameter use korte pari.




plt.title("Count Plot",color="r",fontsize=20)
plt.show()
```
