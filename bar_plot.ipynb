# bar plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

```

```python
var = sns.load_dataset("penguins")
var
```

```python
sns.barplot(x="island",y="bill_length_mm",data=var,hue="sex",err_kws={"color":'y',"linewidth":5},dodge = True,alpha=1)

# amra x=var.island,y=var.bill_length_mm etao use korte pari. ekhetre data=var use korte hobe na.
# hue parameter diye amra ekhane sex er male and female alada kore dekhiyechi. ota na use korle duita eksathe marged thakto.
# ci= 0 to 100 er value change korar maddhome amra bar plot e confidence interval fix korte pari. eta basically bar er mathay jeta thake. 
# n_boot=2 or iccha moto value diye confidence interval er customization kora jay.
# amra orient="h" parameter use kore horizontal graph banate pari. tobe sekhtre duitai numerical value hote hobe. alada alada hole hobe na.
# color="r" erkom kore color change kora jete pare.
# palette="Accent" use koreo colour change kora jabe.
# saturation=value use kore amra saturation fix krte pari.

# errcolor='icchamoto color' diye amra ci er color fix korte pari.
# errwidth=5 eorkom value diye amra ci er width customize korte pari.
# errcolor and errwidth ta amra err_kws{"color":'y',"linewieth":5} er vitor eksathe korte pari. eta better and eta future e use hobe.

# capsize=.5 erokom value diye capsize cutomize korte pari.
# dodge=False kore dile male female each other overlap kore.
# jhapsa korar jonno amra alpha parameter use kori. jar value 0 to 1 obdi.

# amra to matplotlib e grid() use korechi. Seaborn teo grid ache jeta beshi advanced.
# sns.set(style="darkgrid") eta diye grid change kora jay. ekhane darkgrid,dark egula use kora jay. ask chatgpt for more varient.



sns.set(style="darkgrid")




plt.title("Bar Plot",color='r',fontsize=20)
plt.show()
```

```python
# ekhane amra order change korbo.

order_1 =["Biscoe","Dream","Torgersen"] # ei order e hobe.
sns.barplot(x="island",y="bill_length_mm",data=var, hue="sex",order=order_1,hue_order=["Female","Male"])

#hue_order diye amra male female er order o change korte pari. ekhane amra list er use korbo.

plt.show()
```
