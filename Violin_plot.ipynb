# Violine Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np


var = sns.load_dataset("tips")
var
```

```python
sns.violinplot(x="day",y="total_bill",data=var,hue="sex",split=False,inner="quart")


# eta onek ta matplotlib er boxplot er moto
# ekhaneo amra palette, saturation, linewidth egula use korte pari.
# split=True করলে দুটি ক্যাটাগরির ডিস্ট্রিবিউশনকে একটি মাত্র ভায়োলিনের মধ্যে ভাগ করে দেখায়।
# split=False (ডিফল্ট ভ্যালু) হলে আলাদা ভায়োলিন প্লট তৈরি হয় প্রতিটি ক্যাটাগরির জন্য।
# scale="width" parameter diye amra customize korte pari. scale er vitor aro onk kichu deya jay. ask google.
# inner="quart" ভায়োলিন প্লটের ভেতরে 25%, 50%, এবং 75% কোটাইলের লাইন দেখায়, যা ডিস্ট্রিবিউশনের গুরুত্বপূর্ণ পরিসংখ্যান বোঝায়। এটি ডাটা স্কেল ও ছড়িয়ে থাকা ডেটা সহজে বোঝার জন্য উপকারী। 
# inner er vitor box, stick,quart, None use korte pari.
# 


plt.show()
```
