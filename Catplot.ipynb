# Catplot

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

var = sns.load_dataset("tips")
var
```

```python
sns.catplot(x="day",y="total_bill",data=var,kind="boxen",hue="sex")

# catplot er subidha holo eta diye onek dhoroner plot banano jay.
# catplot er vitore amra kind hisebe (strip, swarm, box, violin, boxen, point, bar, count) egula use korte pari.
# color er jonno palette parameter use korte hobe.



plt.show()
```
