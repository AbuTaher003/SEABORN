# Histogram Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

var = sns.load_dataset("penguins")
var
```

```python
sns.displot(var["bill_length_mm"],kde=True,rug=True,color="g",label="bill_length_mm") 

# histogram mainly ekta attribute er jonno kora hoy.

# age distplot() use kora hoto. ekhon eta puraton hoye geche, tobe eta use krleo hobe. halka kore ekta error ashbe arki. 
# ekhon amra histplot() or displot() use korbo. Eta seaborn er updated function.
# bins=[35,40,45,50,55,60] use kore amra range thik korte pari.
# kde = Ture kore dile bar gular upor ekta line create hoy. etake kernal density estimator bole.
# color="g" diye amra icchamoto color change korte pari graph er.
# label use kore amra legend er sahajje label dite pari.

#---------------------------------------------------------------------------------------------------------------------------------
# rug=True kore dile x-axis er upor scaling er moto dag thake. eta kde er path follow kore. ekhane ekta kahini ache.
# histplot() use krle rug use korte ektu jhamela hy. tokhon alada kore sns.rugplot(var["bill_length_mm"],color='r') use korte hy.
# tai amra eto jhamela te na giye displot() use korbo.
#---------------------------------------------------------------------------------------------------------------------------------





plt.legend()
plt.title("Histogram Plot",color='r',fontsize=20)
plt.show()
```
