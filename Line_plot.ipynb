# Line Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

var = [1,2,3,4,5,6,7]
var_1 = [2,3,1,4,5,6,7]

x_1 = pd.DataFrame({"var":var,"var_1":var_1})


sns.lineplot(x="var",y="var_1",data = x_1)  

# sns.lineplot(x="var",y="var_1") just evabe run koralei hobe na cz seaborn dataset niye kaj kore. tai age pandas use kore dataframe banate hobe, sei dataframe ekhane pass kore dilei hobe.
# ekhane data variable ta DataFrame theke pawa data nijer kache rakhe.
# amra ekhane ekta jinish dekhetechi j seaborn e graph er x,y axis er label automatically hoye thake, jeta matplotlib e hoy na. 

plt.show()
```

```python
plt.plot(var,var_1) # ekhane amra matplotlib diye line plot banailam.
plt.show()
```

```python
y_1 = sns.load_dataset("penguins") 

# ekhane amra random ekta dataset niyechi github theke. and seta load korechi ei command diye. seta y_1 variable e store korechi.
# ekhane .head(20) lagay diye dataset ke choto kore niye kaj korte pari.

y_1
```

```python
sns.lineplot(x="bill_length_mm",y="flipper_length_mm",data = y_1,hue="sex",size=50,style="sex",palette = "pastel",markers=["o",">"],legend=True,dashes=False) 


# duita column or property er moddhye amra line plot use korechi. 
# hue parameter dataset e sobtheke powerful data er respect e graph banay.
# hue parameter use korle label automatically add hoy. matplotlib e label er jonno legend() use korte hoy.
# size iccha moto ekta deya jay.
# style parameter er maddhome amra kono attribute er style change kora jay. 
# matplotlib e amra colorbar use kortam, seaborn e amra same jinis pabo palette() use kore. palette er vitor kon color dibo seta jante google korlei hobe.
# markers parameter er maddhme amra graph er sharp corner gula te marker add korte pari. marker amra duita diyechi karon ekhane male and female ache. 
# legend = False kore dile label gula chole jabe. true korle label thakbe.
# dashes = False kore dile dot dot gula dash diye puron hobe. i.g straight lines er moto hbe. ture korle abar dot dot chole ashbe. obosso ei dot dot amra style change korar shomoy peyechi.



plt.title("Line Plot",color='r',fontsize=15) # title add korte pari.
plt.grid() # box box grid add korte pari.
plt.show()
```
