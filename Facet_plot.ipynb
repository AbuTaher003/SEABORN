# Facet Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

var = sns.load_dataset("tips")
var
```

```python
fg = sns.FacetGrid(var,col="day",hue="sex") # egula diye amra graph er jonno plot ready kore rakhlam.. col="day" means daywise alada alada plot create hobe.

fg.map(plt.scatter,"total_bill","tip").add_legend()

# ekhane map er maddhome amra plot e graph set korte pari.. amra ekhane jekono type er graph use korte pari. 
# add_legend() diye amra label add korte pari.


plt.show()
```

```python
# for practice : 

sns.FacetGrid(var,col="time",hue="day").map(sns.lineplot,"total_bill","tip").add_legend()

# scatterplot,lineplot,barplot,boxplot,violinplot,stripplot,swarmplot,kdeplot,histplot egula amra map er vitore use korte parbo.


plt.show()
```
